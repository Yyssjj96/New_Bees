**금일 학습내용 **
  - *데이터 시각화 : 라이브러리 : matplotlib.pyplot*
    - 선 그래프 : plot()
      - 범주 : legend
      - xlim(start, end) 해당 범위로 다시 그리기 (확대)
      - xticks(rotation = 90) x 축 글자를 90도로 회전
    - figure(figsize = (10,5)) 그리려는 가로 세로 폭 조절
    - pit.subplot(행,렬,숫자) : 한 셀에 여러개로 분할해서 그린다.숫자는 그리려는 위치

  - *판다스*
    - 데이터프레임 변수[컬럼명]
    - 데이터프레임 변수[start:end] = row 가져오기
    - loc[행,열] : [문자열,문자열, 문자열:문자열]
    - iloc[행,열] : [인덱스:인덱스, 인덱스:인덱스]

```python
def circle1():
  circle = plt.Circle((0,0), radius = 0.5)
  return circle

def show1(patch):
  ax = plt.gca()
  ax.add_patch(patch)
  plt.axis('scaled')
  plt.show()

if __name__ == '__main__':
  c = circle1()
  show1(c)
```
